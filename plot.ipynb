```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
typ = 'river'
ALG = ['full', 'singlem', 'nouncern', 'hoeffding', 'qlearning', 'random']
num = 10
FINAL = {}
for alg in ALG:
    RESULT = []
    file = './river_h_500/'
    if alg == 'full':
        file += 'full_0.001_1_0.01_0.5_'
    if alg == 'singlem':
        file += 'singlem_0.001_1_0.01_0.5_'
    if alg == 'nouncern':
        file += 'nouncern_0.001_1_0.01_0_'
    if alg == 'hoeffding':
        file += 'hoeffding_0.001_1_0.01_0_'
    if alg == 'random':
        file += 'random_'
    if alg == 'qlearning':
        file += 'qlearning_0.005_'
    for ii in range(num):     
        result = np.loadtxt(file+str(ii))
        RESULT.append(result.reshape(len(result), 1))
    FINAL[alg] = RESULT[0]
    for ii in range(1, len(RESULT)):
        FINAL[alg] = np.concatenate((FINAL[alg], RESULT[ii]), axis = 1)

LABEL = {'full': 'HF-UCRL-VTR+ with M = 4',\
        'singlem': 'HF-UCRL-VTR+ with M = 2',\
        'nouncern': 'UCRL-VTR+',\
        'hoeffding': 'UCRL-VTR',\
        'random': 'Random',\
        'qlearning': 'Q-learning'}
fig, subplot = plt.subplots(1, figsize = (10, 7))
for alg in ALG:
    x = list(range(FINAL[alg].shape[0]))
    y = FINAL[alg].mean(axis = 1)
    std = FINAL[alg].std(axis = 1)
    subplot.plot(x,y, label=LABEL[alg])
    subplot.set_xlabel('number of episodes', fontsize = 20)
    subplot.set_ylabel('average reward', fontsize = 20)
    subplot.set_ylim([0, 42])
    subplot.fill_between(x, y - std, y + std, alpha=0.2)
    subplot.legend(fontsize = 20)
plt.show()
```
